Bubble Sort

```python
from random import randint

V = [randint(0, 14) for _ in range(14)]
n = len(V)
print(f"lista original: {V}")
c = n
for i in range(n-1):
    for j in range(0, c-1):
        if V[j] > V[j+1]:
            temp = V[j]
            V[j] = V[j+1]
            V[j+1] = temp
    c -= 1
print(f"Lista ordenada: {V}")
```
