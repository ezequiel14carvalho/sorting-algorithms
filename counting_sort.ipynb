Counting Sort

```python
from random import randint

def counting(V, n):
    big = 0
    for num in V:
        if num >= big:
            big = num + 1
    count = [0 for x in range(big)]

    for i, _ in enumerate(count):
        c = 0
        for value in V:
            if i == value:
                c += 1
        count[i] = c


    V_sorted = [0 for x in range(n)]
    c = 0
    for i, value in enumerate(count):
        if value > 0:
            for _ in range(value):
                V_sorted[c] = i
                c += 1
    
    return V_sorted


V = [randint(0, 14) for _ in range(14)]
n = len(V)
V_sorted = counting(V, n)

print(f"Lista original: {V}")
print(f"Lista ordenada: {V_sorted}")
```
