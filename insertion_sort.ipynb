Insertion Sort

```python
from random import randint

V = [randint(0, 14) for _ in range(14)]
n = len(V)
print(f"Lista original: {V}")

for i in range(1, n):
    for j in range(i, 0, -1):
        if V[j] < V[j-1]:
            temp = V[j-1]
            V[j-1] = V[j]
            V[j] = temp

print(f"Lista ordenada: {V}")
```
