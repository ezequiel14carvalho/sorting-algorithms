Selection Sort

```python
from random import randint

def selection(V, n):

    for i in range(0, n):
        min_index = i
        for j in range(i+1, n):
            if V[j] < V[min_index]:
                min_index = j

        temp = V[i]
        V[i] = V[min_index]
        V[min_index] = temp

    return V

V = [randint(1, 14) for _ in range(10)]
n = len(V)
V_sorted = selection(V.copy(), n)
print(f"Lista Original: {V}\nLista ordenada: {V_sorted}")
```
