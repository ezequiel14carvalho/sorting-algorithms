Selection Sort

```python
from random import randint

def selection(V, n):

    for i in range(0, n):
        min_index = i
        for j in range(i+1, n):
            if V[j] < V[min_index]:
                min_index = j

        temp = V[i]
        V[i] = V[min_index]
        V[min_index] = temp

    return V

V = [randint(1, 14) for _ in range(10)]
n = len(V)
V_sorted = selection(V.copy(), n)
print(f"Lista Original: {V}\nLista ordenada: {V_sorted}")
```

Bubble Sort

```python
from random import randint

V = [randint(0, 14) for _ in range(14)]
n = len(V)
print(f"lista original: {V}")
c = n
for i in range(n-1):
    for j in range(0, c-1):
        if V[j] > V[j+1]:
            temp = V[j]
            V[j] = V[j+1]
            V[j+1] = temp
    c -= 1
print(f"Lista ordenada: {V}")
```

Insersion Sort

```python
V = [randint(0, 14) for _ in range(14)]
n = len(V)
print(f"Lista original: {V}")

for i in range(1, n):
    for j in range(i, 0, -1):
        if V[j] < V[j-1]:
            temp = V[j-1]
            V[j-1] = V[j]
            V[j] = temp

print(f"Lista ordenada: {V}")
```
